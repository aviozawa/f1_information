```python
# =================================================================
# F1 Pit Optimizer - フェーズ1 & 2 分析ワークフロー
# =================================================================
# 目的：FastF1で取得したデータを基に、タイヤの性能分析と戦略シミュレーションを行う。
# 担当：(あなたの名前)
# 更新日：(今日の日付)
# ---

# --------------------------------------------------
# STEP 1: 準備 (Setup)
# 必要なライブラリと自作モジュールをインポートし、データを読み込む。
# --------------------------------------------------
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("--- [Step 1] セットアップ開始 ---")

# パス設定（Notebookがどこにあっても動作するように）
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
src_path = os.path.join(project_root, 'src')
if src_path not in sys.path:
    sys.path.append(src_path)
    print(f"'{src_path}' をPythonパスに追加しました。")

# 自作の分析関数をインポート
try:
    from analysis import (
        calculate_degradation_per_stint,
        calculate_theoretical_best_lap,
        predict_stint_time
    )
    print("分析モジュールのインポート成功。")
except ImportError as e:
    print(f"分析モジュールのインポートに失敗: {e}")
    # エラーがあればここで処理を止める
    raise

# データ読み込み
DATA_FILE = '2023_japan_r_laps.csv'
file_path = os.path.join(project_root, 'data', DATA_FILE)
try:
    laps_df = pd.read_csv(file_path)
    print(f"データ '{DATA_FILE}' の読み込み成功。データ形状: {laps_df.shape}")
except FileNotFoundError:
    print(f"データファイルが見つかりません: {file_path}")
    print("src/data_fetcher.py を実行してデータを取得してください。")
    raise

print("--- [Step 1] セットアップ完了 ---\n")


# --------------------------------------------------
# STEP 2: 分析 (Analysis)
# 主要ドライバーのデグラデーション率と理論上のベストタイムを計算する。
# --------------------------------------------------
print("--- [Step 2] データ分析開始 ---")

# 分析対象
DRIVER_1 = 'VER'
DRIVER_2 = 'NOR'

# ドライバー1の分析
ver_stint1_laps = laps_df[(laps_df['Driver'] == DRIVER_1) & (laps_df['Stint'] == 1)]
ver_stint1_degradation = calculate_degradation_per_stint(laps_df, DRIVER_1, 1)
theoretical_best_stint1 = calculate_theoretical_best_lap(ver_stint1_laps)

# ドライバー2の分析
nor_stint1_laps = laps_df[(laps_df['Driver'] == DRIVER_2) & (laps_df['Stint'] == 1)]
nor_stint1_degradation = calculate_degradation_per_stint(laps_df, DRIVER_2, 1)
theoretical_best_nor1 = calculate_theoretical_best_lap(nor_stint1_laps)

print(f"[{DRIVER_1}] Stint 1: Degradation={ver_stint1_degradation:.4f} sec/lap, Theoretical Best={theoretical_best_stint1:.3f} sec")
print(f"[{DRIVER_2}] Stint 1: Degradation={nor_stint1_degradation:.4f} sec/lap, Theoretical Best={theoretical_best_nor1:.3f} sec")

print("--- [Step 2] データ分析完了 ---\n")


# --------------------------------------------------
# STEP 3: 予測 (Prediction & Simulation)
# 分析結果を基に、未来のラップタイムをシミュレーションする。
# --------------------------------------------------
print("--- [Step 3] 戦略シミュレーション開始 ---")

# シナリオ設定
START_TYRE_LIFE = 17 # タイヤ年齢17周目から
NUM_LAPS_TO_PREDICT = 5
print(f"シナリオ: タイヤ年齢{START_TYRE_LIFE}周目から、{NUM_LAPS_TO_PREDICT}周走行した場合の合計タイムを予測。")

# ドライバー1の予測
predicted_total_time_ver = predict_stint_time(
    theoretical_best=theoretical_best_stint1,
    degradation_rate=ver_stint1_degradation,
    start_tyre_life=START_TYRE_LIFE,
    num_laps=NUM_LAPS_TO_PREDICT
)

# ドライバー2の予測
predicted_total_time_nor = predict_stint_time(
    theoretical_best=theoretical_best_nor1,
    degradation_rate=nor_stint1_degradation,
    start_tyre_life=START_TYRE_LIFE,
    num_laps=NUM_LAPS_TO_PREDICT
)

print(f"  -> 結果 ({DRIVER_1}): {predicted_total_time_ver:.3f} 秒")
print(f"  -> 結果 ({DRIVER_2}): {predicted_total_time_nor:.3f} 秒")
difference = abs(predicted_total_time_ver - predicted_total_time_nor)
print(f"  => 両者の差: {difference:.3f} 秒")

print("--- [Step 3] 戦略シミュレーション完了 ---")


# --------------------------------------------------
# STEP 4: 可視化 (Visualization)
# 分析の基になったデータをグラフで確認する。
# --------------------------------------------------
print("\n--- [Step 4] データ可視化 ---")

plt.figure(figsize=(15, 8))
sns.set_theme(style="whitegrid")

# ドライバー1のプロット
sns.scatterplot(data=ver_stint1_laps[ver_stint1_laps['IsAccurate']==True], x='TyreLife', y='LapTimeSeconds', label=f'{DRIVER_1} (Actual)', color='blue')
# ドライバー2のプロット
sns.scatterplot(data=nor_stint1_laps[nor_stint1_laps['IsAccurate']==True], x='TyreLife', y='LapTimeSeconds', label=f'{DRIVER_2} (Actual)', color='orange')

plt.title(f'{DRIVER_1} vs {DRIVER_2} - Stint 1 Performance')
plt.xlabel('Tyre Life (Laps)')
plt.ylabel('Lap Time (seconds)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()

print("--- 全てのワークフローが完了しました ---")
```
